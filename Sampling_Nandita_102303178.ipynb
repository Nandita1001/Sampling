```python
import pandas as pd

data = pd.read_csv("Creditcard_data.csv")
data.head()
```

```python
X = data.drop('Class', axis=1)
y = data['Class']
```

```python
from imblearn.under_sampling import RandomUnderSampler, NearMiss
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.combine import SMOTETomek
```

```python
sampling_methods = {
    "Sampling1": RandomUnderSampler(random_state=42),
    "Sampling2": RandomOverSampler(random_state=42),
    "Sampling3": SMOTE(random_state=42),
    "Sampling4": NearMiss(),
    "Sampling5": SMOTETomek(random_state=42)
}

samples = {}

for name, sampler in sampling_methods.items():
    X_res, y_res = sampler.fit_resample(X, y)
    samples[name] = (X_res, y_res)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

models = {
    "M1": LogisticRegression(max_iter=1000),
    "M2": DecisionTreeClassifier(),
    "M3": RandomForestClassifier(),
    "M4": KNeighborsClassifier(),
    "M5": GaussianNB()
}
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

results = []

for samp_name, (X_s, y_s) in samples.items():
    X_train, X_test, y_train, y_test = train_test_split(
        X_s, y_s, test_size=0.3, random_state=42
    )

    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)

        results.append([model_name, samp_name, acc])
```

```python
result_df = pd.DataFrame(
    results, columns=["Model", "Sampling Technique", "Accuracy"]
)

pivot_table = result_df.pivot(
    index="Model", columns="Sampling Technique", values="Accuracy"
)

pivot_table
```

```python
best_sampling = pivot_table.idxmax(axis=1)
best_sampling

```

```python
pivot_table.to_csv("final_accuracy_table.csv")

from google.colab import files
files.download("final_accuracy_table.csv")
```

```python
best_model_table = best_sampling.reset_index()
best_model_table.columns = ["Model", "Best Sampling Technique"]

best_model_table.to_csv("best_model_sampling_table.csv")
files.download("best_model_sampling_table.csv")
```
